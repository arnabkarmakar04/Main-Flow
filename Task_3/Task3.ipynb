```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("Task_3.csv")
df.head()
```

Barplot

```python
var = df.groupby('sex')['tip'].mean().reset_index()

plt.figure(figsize=(6, 4))
sns.barplot(x='sex', y='tip', data=var, palette='bright',hue='sex', width=0.5)
plt.title('Average Tip by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Tip')
plt.tight_layout()
plt.show()
```

```python
var1 = df.groupby('size')['tip'].mean().reset_index()
plt.figure(figsize=(10, 5))
sns.barplot(x='size', y='tip', data= var1, palette='bright', hue='size')
plt.title('Average Tip Amount by Group Size')
plt.xlabel('Group Size')
plt.ylabel('Average Tip Amount')
plt.tight_layout()
plt.show()
```

```python
var2 = df.groupby('smoker')['total_bill'].mean().reset_index()
plt.figure(figsize=(6, 4))
sns.barplot(x='smoker',y='total_bill', data= var2, palette='bright', hue='smoker', width=0.5)
plt.title('Average Total Bill by Smoker Status')
plt.xlabel('Smoker Status')
plt.ylabel('Average Total Bill ($)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='day', y='total_bill', hue='sex', data=df, palette='bright')
plt.title("Total Bill Amount by Day and Gender")
plt.xlabel("Day")
plt.ylabel("Total Bill Amount")
plt.legend(title='Gender')
plt.show()
```

Lineplot

```python
var3 = df.groupby('day')['total_bill'].mean().reset_index()
plt.figure(figsize=(10, 5))
sns.lineplot(x='day', y='total_bill', data=var3, marker='o', color='blue')
plt.title('Average Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Average Total Bill')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x="size", y="tip", data=df, marker='o')
plt.xlabel("Size")
plt.ylabel("Tip")
plt.title("Size vs. Tip")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
var4 = df.groupby('day')['tip'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.lineplot(data=var4, x='day', y='tip', marker='o')
plt.title('Average Tip by Day')
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.grid()
plt.tight_layout()
plt.show()
```
