Importing pandas and creating a dataframe

```python
import pandas as pd

df = pd.read_csv("Task_2.csv")
df.head()
```

```python
df.count()
```

Dropping the duplicate rows (if any)

```python
df.drop_duplicates()
df.head()
```

Checking the missing values

```python
df.isnull().sum()
```

Copying original dataframe into new dataframes for performing multiple operations

```python
df1 = df.copy()
df2 = df.copy()
df3 = df.copy()
df4 = df.copy()
df5 = df.copy()
df6 = df.copy()
```

Filling missing values with:
    1. Constant Values
    2. Forward Fill
    3. Backward Fill
    4. Interpolate
    5. Mean
    6. Median

```python
df1.fillna(0, inplace= True)
df1
```

```python
df2.fillna(method='ffill', inplace= True)
df2
```

```python
df3.fillna(method='bfill', inplace= True)
df3
```

```python
df4.interpolate(inplace= True)
df4
```

Dropping Observation column due to its format which is not supporting for filling NaN values with "Mean & Median" function

```python
df5 = df5.drop('Observation', axis=1)
df6 = df6.drop('Observation', axis=1)
```

```python
df5.fillna(df5.mean(), inplace=True)
df5
```

```python
df6.fillna(df6.median(), inplace=True)
df6
```

Dropping rows which consists any missing values & describing its accurate informations

```python
df.dropna(axis=0, inplace=True)
df.isnull().sum()
```
